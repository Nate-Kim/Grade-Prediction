````python
import matplotlib.pyplot as plt
import seaborn as sns

x = [1, 2, 3, 4]
r2 = [0.1161, -2.2453, -2.4403, -5.0084]
mse = [11.5215, 42.3031, 44.8452, 78.3205]

sns.set_theme(style='white')
sns.set_context("paper")
plt.figure(figsize=(3.5, 2.5))
ax1 = plt.gca()
# R2 on left y
ax1.plot(x, r2, color='gray', marker='o', linewidth=1, markersize=4, label='R²', alpha=0.8)
ax1.set_xlabel('Polynomial Degree', fontsize=8)
ax1.set_ylabel('R²', fontsize=8, color='gray')
ax1.tick_params(axis='y', labelcolor='gray')
ax1.tick_params(axis='both', labelsize=7)
ax1.set_xticks(x)
# MSE on right y
ax2 = ax1.twinx()
ax2.plot(x, mse, color='black', linestyle='--', marker='s', linewidth=1, markersize=4, label='MSE', alpha=0.8)
ax2.set_ylabel('MSE', fontsize=8, color='black')
ax2.tick_params(axis='y', labelcolor='black')
ax2.tick_params(axis='both', labelsize=7)

# Combined legend
lines, labels = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
ax1.legend(lines + lines2, labels + labels2, fontsize=7, loc='upper left', frameon=False)

plt.title('', fontsize=10)
plt.tight_layout()
plt.savefig('poly_regression_degrees.pdf', dpi=300, bbox_inches='tight')
plt.show()
````
